### **IF/ELIF/ELSE**

- you can have a standalone if statement, or an if combined with elif, or an if combined with else, or an if combined with both elif and else.

- if statement is the first check. If it evaluates to true, only the indented code under it will be executed and the remaining elif and else statements will be passed.

- in case the if statement evaluates to false, the indented code under it will be skipped and will jump to the next elif

- if that elif fails, then it skips the code under it and goes to next elif. and so on until else.

- At any point, if any elif evaluates to true, the next elifs and else will be skipped entirely

HENCE, ORDER OF CONDITIONS MATTERS!


### **EXCEPTION HANDLING**

- the try block is executed first
- if try block raises an error, except blocks are checked
- if try block doesn't raise an error, except blocks aren't checked and else block is executed
- finally block is executed in all cases, whether an exception is raised or isn't

https://www.w3schools.com/python/python_ref_exceptions.asp


```python
var1 = 5
var2 = 0

var1/var2
```

```python
var1 = 5
var2 = 0

try:

  var1/var2


except:

  print("operation failed")
```

```python
var1 = 5
var2 = 0

try:

  var1/var2

except Exception as e:

  print("operation failed")
  print(f"the error was {e}")
```

```python
var1 = 5
var2 = 0

try:

  var1/var2


except Exception:

  print("operation failed")


except ZeroDivisionError:

  print("make sure denominator isn't zero")
  var2 += 1


except NameError:

  print("make sure you have defined your variable")
```

```python
var1 = 5
var2 = 0

try:

  var1/var2


except ZeroDivisionError:

  print("make sure denominator isn't zero")
  var2 += 1


except NameError:

  print("make sure you have defined your variable")


except Exception:

  print("operation failed")
```

```python
var1 = 5
var2 = 0

try:

  var1/var3


except ZeroDivisionError:

  print("make sure denominator isn't zero")
  var2 += 1


except NameError:

  print("make sure you have defined your variable")


except Exception:

  print("operation failed")
```

```python
var1 = 5
var2 = 0

try:

  result = var1/var2


except ZeroDivisionError:

  print("make sure denominator isn't zero")
  var2 += 1


except NameError:

  print("make sure you have defined your variable")


except Exception:

  print("operation failed")


else:

  marks = result * 100
  print(f"marks are {marks}")


finally:

  print("program ended")
```

```python
var1 = 5
var2 = 10

try:

  result = var1/var2


except ZeroDivisionError:

  print("make sure denominator isn't zero")
  var2 += 1


except NameError:

  print("make sure you have defined your variable")


except Exception:

  print("operation failed")


else:

  marks = result * 100
  print(f"marks are {marks}")


finally:

  print("program ended")
```

```python
def add2(collection):

  total = 0

  try:

    for i in collection:

      total += i

  except TypeError:

    print("make sure you only have numbers in your list")

  else:

    return total
```

```python
add2((1,2,3,6,-4,10))
```

```python
add2((1,2,3,"b",-4,10))
```

```python
def add2(collection):

  total = 0

  for i in collection:

    try:

      total += i

    except TypeError:

      continue


  return total
```

```python
add2((1,2,3,6,-4,10))
```

```python
add2((1,2,3,"b",-4,10))
```

```python
add2((1,2,3,"b",-4,10,[1,2,3]))
```

```python
add2((1,2,3,-4,10,True))
```
